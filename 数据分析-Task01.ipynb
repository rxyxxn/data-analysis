### 任务一：pandas中有两个数据类型DateFrame和Series，通过查找简单了解他们。然后自己写一个关于这两个数据类型的小例子

（1）Series 一维，带标签数组,与Numpy中的一维array类似。    
    <font size="3" color="red">由两个数组构成:    
    构成对象的键（index，索引），一个数组构成对象的值（values），键——>值</font>

#### Series创建

- 通过列表创建     
pd.Series(列表，index=....)   ——>index指定元素的索引

```python
t = pd.Series(np.arange(10), index=[i for i in range(10)])
```

```python
t
```

```python
type(t)
```

- 通过字典创建    
pd.Series(字典)   ——>使用字典中原数据的键值

```python
t = pd.Series({"name":"hanmh-", "age":"21", "language":"chinese"})
```

```python
print(t)
type(t)
```

- 字典推导式创建字典，并作为Series的参数

```python
import string
a = {string.ascii_uppercase[i]:i for i in range(10)}
t = pd.Series(a)

print(a)
print(t)
```

- 重新给上面字典指定其他索引后，如果能够对上，就取其值，如果不能就直接置为nan

```python
t = pd.Series(a, index=list(string.ascii_uppercase[5:15]))
t
```

- 修改Series的dtype值     

t.astype(类型)

- pandas之Series切片和索引     

切片：直接传入start end或者步长即可    

索引：一个的时候直接传入序号或者index，多个的时候传入序号或者index的列表

- 取出Series的索引或者值

t.index      取出Series的索引   

t.values     取出Series中具体的值


- Series中的where方法     
该方法与numpy中的where方法输出结果不一样，pandas中的where是输出匹配项，不匹配的直接赋值为nan

```python
a = {string.ascii_uppercase[i]:i for i in range(10)}
t = pd.Series(a)
print(t)
t = t.where(t>5)
print(t)
```

#### 缺失数据处理

dropna() # 过滤掉值为NaN的行    
fillna() # 填充缺失数据    
isnull() # 返回布尔数组，缺失值对应为True     
notnull() # 返回布尔数组，缺失值对应为False

```python
t.dropna() 
```

```python
t.fillna(1)
```

(2)DataFrame是一个表格型的数据结构，相当于是一个二维数组，含有一组有序的列。他可以被看做是由Series组成的字典，并且共用一个索引（这里的索引指的就是列索引）。

#### 创建方式

- 第一种：# 产生的DataFrame会自动为Series分配所索引，并且列会按照排序的顺序排列

```python
d = pd.DataFrame({'one':[1,2,3,4],'two':[4,3,2,1]})
```

```python
d
```

- 指定列   
可以通过columns参数指定顺序排列

```python
pd.DataFrame(d,columns=['two','one'])  #指定显示顺序
```

```python
#第二种：
pd.DataFrame({'one':pd.Series([1,2,3],index=['a','b','c']),'two':pd.Series([1,2,3],index=['b','a','c'])})
```

- 常用属性和方法：（和numpy类似）   

index 获取行索引    
columns 获取列索引   
T 转置    
values 获取值    
describe() 获取快速统计
